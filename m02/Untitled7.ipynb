```python
df = pd.read_csv("drive/MyDrive/jksl_housing_price.csv")
df.head(11)
```

```python
# membuat baris pertama (0) menjadi header dan hapus di dari baris 0
df.columns = df.iloc[0]
df = df.drop(df.index[0])
```

```python
df.info()
```

```python
# Hapus data yang tidak diperlukan
df = df.drop(['GRS', 'KOTA'], axis = 1)
```

```python
# Cek
df.info()
df.head(11)
```

```python
df["HARGA"] = pd.to_numeric(df["HARGA"])
df["LT"] = pd.to_numeric(df["LT"])

harga = df["HARGA"]
lt = df["LT"]

# Menghitung nilai interpolasi linear dengan rumus
luastanah_6 = lt[5] + (harga[6] - harga[5]) * (lt[5] - lt[7]) / (harga[5] - harga[7])

# print hasil
print("LT (Luas Tanah) pada data ke-6:", luastanah_6)
```

```python
df["LT"].fillna(luastanah_6, inplace=True)
```

```python
df.head(11)
```

```python
# Menghitung rata-rata dari 'HARGA' dan 'LT'
rata_harga = np.mean(harga)
rata_lt = np.mean(lt)

# Rumusnya adalah Y = a + bX
# Menghitung slope (b)
numerator = sum((harga - rata_harga) * (lt - rata_lt))
denominator = sum((harga - rata_harga)**2)
slope = numerator / denominator

# Menghitung intercept (b)
intercept = rata_lt - slope * rata_harga

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")

# Misalkan ingin mengekstrapolasi nilai LT untuk harga baru
harga_baru = 3000000000  # contoh harga baru

# Menghitung ekstrapolasi LT untuk harga baru
lt_ekstrapolasi = slope * harga_baru + intercept

print(f"LT (Luas Tanah) untuk harga baru {harga_baru} adalah: {lt_ekstrapolasi}")
```
